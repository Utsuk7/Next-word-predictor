```python
with open('/content/1661-0.txt') as f:
  data = f.read()
```

```python
#print(data)
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import tensorflow as tf
from tensorflow.keras.preprocessing.text import Tokenizer
from tensorflow.keras.preprocessing.sequence import pad_sequences
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Embedding, LSTM, Dense
```

```python
tokenizer=Tokenizer()
```

```python
tokenizer.fit_on_texts([data])
```

```python
len(tokenizer.word_index)
```

```python
input_sequence=[]
for sentence in data.split('\n'):
  tokenizers=tokenizer.texts_to_sequences([sentence])[0]

  for i in range(1,len(tokenizers)):
    n_gram_sequence=tokenizers[:i+1]
    input_sequence.append(n_gram_sequence)
```

```python
input_sequence
```

```python
max_len = max([len(x) for x in input_sequence])
max_len
```

```python
padded_input_sequence = pad_sequences(input_sequence, maxlen=max_len, padding='pre')
```

```python
padded_input_sequence
```

```python
X=padded_input_sequence[:,:-1]
```

```python
y=padded_input_sequence[:,-1]
```

```python
print(X.shape)
print(y.shape)
```

```python
len(tokenizer.word_index)
```

```python
from tensorflow.keras.utils import to_categorical
y=to_categorical(y,num_classes=len(tokenizer.word_index)+1)
```

```python
y.shape
```

```python
model=Sequential()
model.add(Embedding(input_dim=len(tokenizer.word_index)+1,output_dim=100,input_length=max_len-1))
model.add(LSTM(100))
model.add(Dense(len(tokenizer.word_index)+1,activation='softmax'))
```

```python
model.compile(loss='categorical_crossentropy',optimizer='adam',metrics=['accuracy'])
```

```python
model.summary()
```

```python
model.fit(X,y,epochs=100)
```

```python
text='my name'
for i in range(10):
  token_list=tokenizer.texts_to_sequences([text])[0]

  padded=pad_sequences([token_list],maxlen=max_len-1,padding='pre')
  position=np.argmax(model.predict(padded))
  #print(position)
  for word,index in tokenizer.word_index.items():
    if index==position:
      text=text+ " " +word
      print(text)
```
